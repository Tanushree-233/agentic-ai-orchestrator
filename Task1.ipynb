```python
import asyncio
import time

class PlannerAgent:

    def plan(self, task):

        return [
            "Retrieve Information",
            "Analyze Information",
            "Write Report"
        ]


class RetrieverAgent:

    def retrieve(self, query):

        if query == "FAIL":
            raise Exception("Retriever Failed")

        return """
Applications:
Disease prediction
Medical imaging

Advantages:
Fast diagnosis

Challenges:
Privacy concerns
"""


class AnalyzerAgent:

    def analyze(self, text):

        return {
            "summary": text
        }


class WriterAgent:

    def write(self, analysis):

        return f"""
FINAL REPORT

{analysis['summary']}
"""


async def stream(msg):

    print(msg)
    await asyncio.sleep(1)


async def main():

    task = "AI in Healthcare"

    planner = PlannerAgent()
    retriever = RetrieverAgent()
    analyzer = AnalyzerAgent()
    writer = WriterAgent()

    await stream("Planner Agent Started")

    steps = planner.plan(task)

    print("Steps:")
    print(steps)

    try:

        await stream("Retriever Agent Started")

        data = retriever.retrieve(task)

        await stream("Retriever Completed")

    except Exception:

        print("Retriever Failed")
        data = "Fallback Data"

    await stream("Analyzer Agent Started")

    analysis = analyzer.analyze(data)

    await stream("Analyzer Completed")

    await stream("Writer Agent Started")

    report = writer.write(analysis)

    await stream("Writer Completed")

    print("\n")
    print(report)



await main()
```
